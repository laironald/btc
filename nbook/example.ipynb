```python
import time

mtgox = open("btc/mtgoxUSD.csv", "rb")
time_interval = 60*10
time_initial = 1279408000
marks = 0
min_mark = 0
max_mark = 0
close_mark = 0
total_volume = 0.0
count = 0
mark_point = 0
break_point = 10000

print "{:5} {:10} {:8} {:20} {:12} {:12} {:12} {:12}".format(
    "count", "point", "marks", "time", "min_mark", "max_mark", "close_mark",
    "volume")
print "-"*100

for i,row in enumerate(mtgox):
    ctime, mark, volume = row[:-1].split(",")
    ctime, mark, volume = (int(ctime), float(mark), float(volume))
    point = (ctime - time_interval) / time_interval
    if point > mark_point:
        marks += 1
        count += 1
        print "{:5} {:10} {:8} {:20} {:12.5f} {:12.5f} {:12.5f} {:12.2f}".format(
            count, mark_point, marks, time.strftime('%Y-%m-%d %H:%M:%S', time.localtime(ctime)), 
            min_mark, max_mark, close_mark, total_volume)
        mark_point = (ctime - time_interval) / time_interval
        min_mark = mark
        max_mark = mark
        close_mark = mark
        total_volume = volume
        marks = 0
    else:
        min_mark = min(mark, min_mark)
        max_mark = max(mark, max_mark)
        total_volume += volume
        marks += 1
        close_mark = mark
        
    if i == break_point - 1:
        break

```
